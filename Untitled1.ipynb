```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cvxpy as cp
from scipy.optimize import minimize
from scipy.optimize import linprog
```

```python
#punto 1
# Definir las funciones objetivo y las restricciones
c = np.array([-3, -5])
A = np.array([[1, 1], [2, 4], [1, 2]])
b = np.array([500, 1000, 2000])
bounds = ((100, None), (150, None))

# Resolver el problema de programación lineal
res = linprog(c=c, A_ub=A, b_ub=b, bounds=bounds)

# Imprimir los resultados
print('Ganancias máximas: $' + str(-res.fun))
print('Cantidad de unidades producidas de A: ' + str(res.x[0]))
print('Cantidad de unidades producidas de B: ' + str(res.x[1]))
```
